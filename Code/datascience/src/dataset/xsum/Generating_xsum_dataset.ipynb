```python
from datasets import load_dataset, load_metric

raw_datasets = load_dataset("xsum")
```

```python
print(raw_datasets)
```

```python
raw_datasets['train'][1]
```

```python
data=[]
```

```python
for i in range(2000):
  data.append({'document': raw_datasets['train'][i]['document'] , 'summary': raw_datasets['train'][i]['summary']})
```

```python
len(data)
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.to_csv('xsum.csv', index=False)
```
